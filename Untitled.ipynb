```python
from nn import NeuralNetwork
from layer import Layer
import pandas as pd
import swifter
import numpy as np
```

```python
df = pd.read_csv('mnist_train.csv')
```

```python
df.head(4)
```

```python
def cols_to_img(row):
    return np.array(row, dtype=float)[1:(28*28 + 1)]
def label_to_vec(label):
    arr = np.zeros(10, dtype=float)
    arr[int(label)] = 1.
    return arr
```

```python
df['img'] = df.swifter.apply(cols_to_img, axis=1)
df['out'] = df['label'].swifter.apply(label_to_vec)
```

```python
df['out_t'] = df['out'].apply(lambda x : np.array([x]).T)
df['img_t'] = df['img'].apply(lambda x : np.array([x]).T)
```

```python
nn = NeuralNetwork([
    Layer(28 * 28, 10, "identity")
])
```

```python
nn.train(
    df, in_col = 'img_t', out_col = 'out_t',
    batch_size = 16, alpha = 0.00001,
    stopper = lambda err, i : i == 1000
)
```
